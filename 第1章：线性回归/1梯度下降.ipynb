```python
import numpy as np
import matplotlib.pyplot as plt
```

求y=x^2+2x+5的最小值

```python
x=np.linspace(-6,4,100)
y=x**2+2*x+5
plt.plot(x,y)
```

初始化横坐标与学习率（迭代次数）

```python
x=4
lr=0.1
iteraterNum = 100000
```

迭代求thera

```python
for i in range(iteraterNum):
    gradient = 2*x+2
    x = x-lr*gradient
```

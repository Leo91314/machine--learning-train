```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

加载数据

```python
def load_data():
    data = np.loadtxt('data.txt',delimiter=' ')
    n = data.shape[1]-1
    X = data[:,0:n]
    y = data[:,-1].reshape(-1,1)
    return X,y
X,y = load_data()
X,y
```

1.最小二乘法

```python
model1 = linear_model.LinearRegression()
```

```python
model1.fit(X,y)#使用训练模型
```

```python
print(model1.coef_)#输出系数
print(model1.intercept_)#输出截距

plt.scatter(X[:,0],y)
plt.plot(X[:,0],model1.predict(X),color='red')
```

2.Ridge回归

```python
model2 = linear_model.Ridge(alpha=0.5)
model2.fit(X,y)
```

```python
print(model2.coef_)
print(model2.intercept_)
plt.scatter(X[:,0],y)
plt.plot(X[:,0],model2.predict(X),color='red')
```

3.LASSO回归

```python
model3 = linear_model.Lasso(alpha=0.5)
model3.fit(X,y)
print(model3.coef_)
print(model3.intercept_)
plt.scatter(X[:,0],y)
plt.plot(X[:,0],model3.predict(X),color='red')
```

4.ElactisNet
弹性网模型
结合Ridge和Lasso（L1与L2正则化）

```python
model4 = linear_model.ElasticNet(alpha=0.5)
model4.fit(X,y)
print(model4.coef_)
print(model4.intercept_)
plt.scatter(X[:,0],y)
plt.plot(X[:,0],model4.predict(X),color='red')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import scipy.io
from sklearn.decomposition import PCA
```

```python
def loaddata():
    data = np.loadtxt('data/pca_data.csv', delimiter=',')
    return  data
```

```python
X = loaddata()
model = PCA(n_components=1)
Z = model.fit_transform(X)
print(model.explained_variance_ratio_)#贡献率
print(model.singular_values_)#奇异值
```

```python
X_rec = model.inverse_transform(Z)
plt.scatter(X[:,0], X[:,1], c='r', marker='o', alpha=0.5)
plt.scatter(X_rec[:,0], X_rec[:,1], c='b', marker='x', alpha=0.5)
```

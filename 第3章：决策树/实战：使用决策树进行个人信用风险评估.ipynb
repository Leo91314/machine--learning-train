# 划分训练集于测试集

```python
credit.savings.value_counts()
```

```python
from sklearn import model_selection
y = credit['default']
X = credit.loc[:,'account_check_status':'foreign_worker']
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
print(y_train.value_counts()/len(y_train))
print(y_test.value_counts()/len(y_test))
```

```python
# 创建模型
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
model = DecisionTreeClassifier(criterion='entropy')
g_search = GridSearchCV(model, param_grid={'max_depth':range(1,10),'min_samples_leaf':range(1,10),'min_samples_split':range(2,10)},cv=10)
g_search.fit(X_train, y_train)
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
credit_model = DecisionTreeClassifier(criterion='entropy', max_depth=9, min_samples_leaf=9,min_samples_split=4)
credit_model.fit(X_train, y_train)
```

```python
credit_pred = credit_model.predict(X_test)
from sklearn.metrics import confusion_matrix
print(credit_model.score(X_test, y_test))
print(confusion_matrix(y_test, credit_pred))
```


#  模型性能提升
在实际应用中，模型的预测正确率不高，很难将其应用到实时的信贷评审过程。
在本案例中，如果一个模型将所有的贷款都预测为“未违约”，此时模型的正确率将为72%，而该模型是一个完全无用的模型。
上节中我们建立的模型，正确率为70%，但是对于违约贷款的识别性能很差。
我们可以通过创建一个代价矩阵定义模型犯不同错误时的代价。
假设我们认为一个贷款违约者给银行带来的损失是银行错过一个不违约的贷款带来损失的4倍，则未违约和违约的代价权重可以定义为：

```python
import sklearn
class_weights = {0:1, 1:4}
credit_model_cost = DecisionTreeClassifier(max_depth=6,class_weight = class_weights)
credit_model_cost.fit(X_train, y_train)
credit_pred_cost = credit_model_cost.predict(X_test)

print (sklearn.metrics.classification_report(y_test, credit_pred_cost))
print (sklearn.metrics.confusion_matrix(y_test, credit_pred_cost))
```

# 分类模型评价指标详解：Accuracy、Precision、Recall 与 Fβ-score

在分类任务中，仅用“准确率”（Accuracy）来衡量模型性能可能不够全面。为了更深入地评估模型的表现，我们引入以下四个核心指标：

- Accuracy（正确率）
- Precision（精确度）
- Recall（召回率）
- F<sub>β</sub>-score（Fβ 分数）

本文将逐一解释这些指标的含义、计算方式及适用场景。

---

## 1️⃣ Accuracy（正确率）

### 🧠 含义
表示模型预测 **正确的样本数量** 占 **总样本数量** 的比例。

### 📐 公式
$$
\text{Accuracy} = \frac{TP + TN}{TP + FP + TN + FN}
$$

其中：

- **TP（True Positive）**：将正类预测为正类的数量
- **TN（True Negative）**：将负类预测为负类的数量
- **FP（False Positive）**：将负类预测为正类的数量
- **FN（False Negative）**：将正类预测为负类的数量

### ✅ 优点
- 简单直观，适合样本类别均衡的情况

### ⚠️ 缺点
- 在类别极度不平衡的情况下可能误导。例如：
  - 预测肿瘤患者中 99% 为健康人，Accuracy 可能很高，但模型实际毫无价值。

---

## 2️⃣ Precision（精确度）

### 🧠 含义
预测为正类的样本中，有多少是真的正类。

### 📐 公式
$$
\text{Precision} = \frac{TP}{TP + FP}
$$

### 🎯 示例
如果模型预测了 100 个是正类，其中 80 个是正确的（TP），20 个是错误的（FP），那么：
$$
\text{Precision} = \frac{80}{80 + 20} = 0.8
$$

### ✅ 适用场景
- 当**误报代价高**时，比如垃圾邮件识别、金融欺诈检测等。

---

## 3️⃣ Recall（召回率）

### 🧠 含义
所有真实正类样本中，有多少被模型成功预测为正类。

### 📐 公式
$$
\text{Recall} = \frac{TP}{TP + FN}
$$

### 🎯 示例
如果实际有 100 个正样本，但模型只找出其中 60 个（FN=40），那么：
$$
\text{Recall} = \frac{60}{60 + 40} = 0.6
$$

### ✅ 适用场景
- 当**漏报代价高**时，比如疾病筛查、安全系统、刑侦预测等。

---

## 4️⃣ F<sub>β</sub>-score：精确度与召回率的调和平均

### 🧠 含义
F-score 是 Precision 和 Recall 的综合指标，是它们的**调和平均**，用来在两者之间取得平衡。

### 📐 常见公式（F1-score）：
$$
F_1 = 2 \cdot \frac{Precision \cdot Recall}{Precision + Recall}
$$

### 🧮 F<sub>β</sub>-score 的推广形式：
$$
F_\beta = (1 + \beta^2) \cdot \frac{Precision \cdot Recall}{\beta^2 \cdot Precision + Recall}
$$

- **β > 1**：更看重 Recall（召回率）
- **β < 1**：更看重 Precision（精确度）
- **β = 1**：F1-score，最常用，表示两者同等重要

### 🎯 示例
如果：
- Precision = 0.8
- Recall = 0.6

则：
$$
F_1 = \frac{2 \cdot 0.8 \cdot 0.6}{0.8 + 0.6} = 0.6857
$$

---

## 🔚 总结对比表

| 指标         | 含义                         | 适用场景                         |
|--------------|------------------------------|----------------------------------|
| Accuracy      | 所有预测正确的样本比例       | 类别分布均衡                    |
| Precision     | 预测为正的样本中有多少是真的 | 注重少误报（False Positive）    |
| Recall        | 所有正样本中被预测出来的比例 | 注重少漏报（False Negative）    |
| F1-score      | 精确度与召回率的平衡值       | 综合考虑 Precision 与 Recall    |

---

## 📌 小贴士

- 在二分类任务中，需指定哪个类是“正类（Positive）”。
- 多分类任务中，可使用 `macro`, `micro`, `weighted` 等方式对指标进行平均。
- 用 `sklearn.metrics.classification_report` 可快速生成这些指标。
